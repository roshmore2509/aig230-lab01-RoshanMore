# Matrices and Nested Lists

```python
# In Python, matrices are often represented as lists of lists.
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(matrix)
print(matrix[0])
print(matrix[1][2])  # row 2, column 3
```

### Looping Through a Matrix

```python
for row in matrix:
    for value in row:
        print(value, end=" ")
    print()
```

# Conditional Statements (if / elif / else)

```python
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
else:
    grade = "C"

print("Grade:", grade)
```

# Classes and Objects

```python
# Classes allow you to model real-world objects.
class Student:
    def __init__(self, name, course):
        self.name = name
        self.course = course

    def introduce(self):
        return f"My name is {self.name} and I am enrolled in {self.course}"
```

```python
student1 = Student("Jessica", "AIG 230")
print(student1.introduce())
```

```python
# TODO
""" Create three instances of the courses_winter 2026 class and display their info."""

class Courses_Winter_2026:
    def __init__(self, course_name, code, time, grade):
        self.name = "Roshan"
        self.course_name = course_name
        self.code = code
        self.time = time
        self.grade = grade

    def display_info(self):
        print(f"Student: {self.name}")
        print(f"Course:  {self.course_name} ({self.code})")
        print(f"Time:    {self.time}")
        print(f"Grade:   {self.grade}")
        print("-" * 30)

course1 = Courses_Winter_2026("Natural Language Processing", "AIG 230", "Monday 08:00AM", "TBD")
course2 = Courses_Winter_2026("Deep Learning", "AIG 240", "Wednesday 10:00AM", "TBD")
course3 = Courses_Winter_2026("Computer Vision", "AIG 250", "Friday 01:00PM", "TBD")

course1.display_info()
course2.display_info()
course3.display_info()
```
